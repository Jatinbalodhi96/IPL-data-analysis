```python
matches_df = pd.read_csv('data/matches.csv', encoding='utf-8')
deliveries_df = pd.read_csv('data/deliveries.csv', encoding='utf-8')
```

## Teams with maximum toss won

```python
# plt.figure(figsize=(14,6))
plt.xticks(rotation='vertical')
temp = matches_df.groupby('toss_winner').agg({'toss_winner': 'count'})\
        .sort_values(by='toss_winner', ascending=False)
sns.countplot(x='toss_winner', palette='GnBu_d', data=matches_df[['toss_winner']])
```

```python
plt.xticks(rotation='vertical')
sns.countplot(x='toss_winner',hue='toss_decision', data=matches_df[['toss_winner', 'toss_decision']])
```

```python
plt.xticks(rotation='vertical')
sns.countplot(x='toss_winner',hue='toss_decision', data=matches_df[['venue', 'toss_decision', 'toss_winner']])
```

```python
plt.xticks(rotation='vertical')
sns.countplot(x='venue',hue='toss_decision', data=matches_df[['venue', 'toss_decision', 'toss_winner']])
```
